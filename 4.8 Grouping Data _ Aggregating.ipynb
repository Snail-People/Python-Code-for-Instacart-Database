```python
ords_prods_merge = pd.read_pickle(os.path.join(path, 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
# Create a subset of the ords_prods_merge dataframe

df = ords_prods_merge[:1000000]
```

```python
# Check first 10 rows

df.head(10)
```

```python
# Remove unnecessary column

df = df.drop('Unnamed: 0', axis=1)
df.head(10)
```

```python
# Group the 'product_name' column Group the 'product_name' column

df.groupby('product_name')
```

```python
# Calculate the mean of the 'order_number' column grouped by the 'department_id' column
# Split the data into groups by 'derpartment_id', then aggregate the mean of the 'order_number' column

df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Alternative way to aggregate using .mean()

df.groupby('department_id')['order_number'].mean()
```

```python
# Perform multiple aggregations

df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# Create new column called 'max_order'
# use the groupby function to group user_id and apply the transform order_number with the np.max argument
# np.max is to find the highest number of orders

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Create flag to assign a 'loyalty' label to a user ID

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
