#### Create a function to label products based on price

```python
# Use only the first 1 million rows to avoid any potential issues
df = ords_prods_merged[:1000000]
```

```python
# Double check that the dataframe is the right size
#df.shape
df.head()
```

```python
# Define the function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check the most expensive product within the subset
df['prices'].max()
```

#### Use if-then statements to label products

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

#### Use a for-loop to label the busiest days of the week

```python
# Count which days had the most orders
ords_prods_merged['order_dow'].value_counts(dropna = False)
```

```python
# Day 0 is the busiest and day 4 is the least busiest. Everything else inbetween can be labeeled 'regularly busy'

result = []

for value in ords_prods_merged["order_dow"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

#### Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
days = []

for value in ords_prods_merged["order_dow"]:
  if value == 0:
    days.append("Busiest days")
  elif value == 1:
    days.append("Busiest days")
  elif value == 3:
    days.append("Least busy")
  elif value == 4:
    days.append("Least busy")
  else:
    days.append("Regularly busy")
```

#### When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# I used the top 6 and bottom 6 hours in terms of order numbers to use as labels and everything else can go inbetween

busy_hours = [10, 11, 14, 15, 13, 12]
slow_hours = [3, 4, 2, 5, 1, 0, 6]
hours = []

for hour in ords_prods_merged['order_hour_of_day']:
    if hour in busy_hours:
        hours.append("Most orders")
    elif hour in slow_hours:
        hours.append("Fewest orders")
    else:
        hours.append("Average orders")
```
